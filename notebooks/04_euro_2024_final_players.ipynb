```python
from statsbombpy import sb
import pandas as pd

MATCH_ID = 3943043

lineups = sb.lineups(match_id=MATCH_ID)
```

```python
players = []

for team_name, lineup_df in lineups.items():
    temp = lineup_df.copy()
    temp["team"] = team_name
    players.append(temp)

players_df = pd.concat(players, ignore_index=True)

players_df.head()
```

```python
players_df.columns.tolist()
```

```python
players_df["has_played"] = players_df["positions"].apply(
    lambda x: isinstance(x, list) and len(x) > 0
)

players_played = players_df[
    players_df["has_played"]
].copy()

players_played
```

```python
players_played[
    [
        "team",
        "player_name",
        "jersey_number",
        "country",
        "positions"
    ]
]
```

```python
def extract_main_position(positions):
    if isinstance(positions, list) and len(positions) > 0:
        return positions[0].get("position")
    return None
```

```python
players_played["main_position"] = players_played["positions"].apply(
    extract_main_position
)
```

```python
players_played_clean = players_played[
    [
        "team",
        "player_id",
        "player_name",
        "jersey_number",
        "main_position"
    ]
].sort_values(
    by=["team", "jersey_number"]
)

players_played_clean
```

```python
centre_backs = players_played_clean[
    players_played_clean["main_position"].str.contains(
        "Center Back",
        case=False,
        na=False
    )
].copy()

centre_backs
```

```python
events = sb.events(match_id=MATCH_ID)

event_players = (
    events[["team", "player", "position"]]
    .dropna()
    .drop_duplicates()
    .sort_values(["team", "player"])
)

event_players
```

```python
players_for_configs = players_played_clean.copy()

players_for_configs["match_id"] = 3943043
players_for_configs["competition_id"] = 55
players_for_configs["season_id"] = 282

players_for_configs = players_for_configs[
    [
        "competition_id",
        "season_id",
        "match_id",
        "team",
        "player_id",
        "player_name",
        "jersey_number",
        "main_position"
    ]
]

players_for_configs
```

```python
from pathlib import Path

output_dir = Path("../outputs/players")
output_dir.mkdir(parents=True, exist_ok=True)

players_for_configs.to_csv(
    output_dir / "euro_2024_final_players.csv",
    index=False
)
```
